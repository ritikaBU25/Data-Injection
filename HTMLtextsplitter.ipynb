```python
from langchain_text_splitters import HTMLHeaderTextSplitter
html_string = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Compliance Policy</title>
</head>
<body>

<h1>Information Security Policy</h1>
<p>
This policy defines how sensitive information must be handled across the organization.
</p>

<h2>Data Classification</h2>
<p>
All data must be classified as Public, Internal, Confidential, or Restricted.
</p>

<h3>Confidential Data</h3>
<p>
Confidential data includes customer PII, financial records, and credentials.
</p>

<h2>Access Control</h2>
<p>
Access to systems must follow the principle of least privilege.
</p>

<h3>User Authentication</h3>
<p>
Users must authenticate using multi-factor authentication.
</p>

<h3>Authorization</h3>
<p>
Role-based access control (RBAC) must be enforced.
</p>

<h2>Audit and Logging</h2>
<p>
All access and security events must be logged and retained for 1 year.
</p>

</body>
</html>
"""

headers_to_split_on=[
    ('h1',"Header 1"),
    ('h2',"Header 2"),
    ('h3',"Header 3")
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
html_header_splits=html_splitter.split_text(html_string)
html_header_splits
```

```python
url="https://www.privacyshield.gov/article?id=Privacy-Shield-Principles"
headers_to_split_on=[
    ('h1', "Header 1"),
    ('h2', "Header 2"),
    ('h3', "Header 3")
]
html_splitter=HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
html_header_splits=html_splitter.split_text_from_url(url)
html_header_splits
```
